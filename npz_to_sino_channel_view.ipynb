```python
import numpy as np
import matplotlib.pyplot as plt

# Load the .npz file
file_path = "/Users/safaiathossain/PHD Research/CT_Recon/xray_dataset/xray_projections.npz"
data = np.load(file_path)

# Check available keys
print("Keys in the dataset:", list(data.keys()))

# Extract the projections (update the key based on actual structure)
projections = data[list(data.keys())[0]]  # Load the first key's data

# Print shape to understand dimensions
print("Shape of projections:", projections.shape)  # Example: (num_channels, num_rows, num_columns)

# Select a middle row for visualization
row_index = projections.shape[1] // 2  # Adjust for row dimension
sinogram = projections[:, row_index, :]  # Extracting sinogram along row dimension

# Display the sinogram
plt.figure(figsize=(10, 6))
plt.imshow(sinogram, cmap='gray', aspect='auto')
plt.colorbar(label="Intensity")
plt.xlabel("Projection Angle")
plt.ylabel("Channel Index")
plt.title(f"Sinogram Projection at Row {row_index}")
plt.show()
```
